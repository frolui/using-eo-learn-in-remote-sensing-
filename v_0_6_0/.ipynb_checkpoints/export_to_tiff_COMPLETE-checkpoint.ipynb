```python
# Импорт базовых компонентов, обеспечивающих работу системы

from eolearn.core.eoworkflow import LinearWorkflow
from eolearn.core.eodata import FeatureType
from eolearn.core import LoadTask, EOPatch
from eolearn.io import ExportToTiff

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
patch = EOPatch.load('./result_patch')
```

```python
patch
```

```python
plt.figure(figsize=(10,10))
plt.imshow(patch.data['dNBR'][0].squeeze());

plt.subplots_adjust(bottom=0.1, right=0.8, top=0.9)
cax = plt.axes([0.85, 0.1, 0.075, 0.8])
plt.colorbar(cax=cax)
plt.show()
```

```python
load = LoadTask('./result_patch')

export_tiff = ExportToTiff((FeatureType.DATA, 'dNBR'), folder = './predicted_tiff/dNBR.tiff')
```

```python
workflow_export_D = LinearWorkflow(
    load,
    export_tiff
)
```

```python
result = workflow_export_D.execute()
```
